```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
m1=fits.open('final_output_Q0.lc')
m2=fits.open('final_output_Q1.lc')
m3=fits.open('final_output_Q2.lc')
m4=fits.open('final_output_Q3.lc')
print(m1.info())
```

```python
m1[0].header
plt.figure(figsize=(10,6))
```

```python
(m1[1].header)
```

```python
m1_data = m1[1].data
print("Required Data:")
print(m1_data)
print(m1_data.shape, m1_data.dtype)
```

```python
m_complete = [m1,m2,m3,m4]
for m in m_complete:
    plt.subplots(1, 1, figsize = (15,9))
    plt.plot(m[1].data['TIME'],m[1].data['RATE'])
    plt.xlabel("Time")
    plt.ylabel("Count Rate")
    
```
